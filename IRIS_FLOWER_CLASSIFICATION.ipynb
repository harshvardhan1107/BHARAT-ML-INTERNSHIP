```python
# Importing crucial library
import pandas as pd
# importing dataset from sklearn
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris) # to get the info
```

```python
iris.data # to check the data
```

```python
iris.feature_names  # to print the name of different feature
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names) # to define the data frame
df.head()
```

```python
df['target'] = iris.target # to set the target variable
df.head()
```

```python
iris.target_names # showing various categories
```

```python
#df[df.target==0].head()
#df[df.target==1].head()
df[df.target==2].head()  #printing the rows of different attribues
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])  # to define the name of different flowers
df.head()
```

```python
from matplotlib import pyplot as plt # importing for visualisation
%matplotlib inline
```

```python
df0 = df[df.target ==0]  # defining the target
df1 = df[df.target ==1]
df2 = df[df.target ==2]
```

```python
df2.head() # printing 5 rows data from the dataset
```

```python
# ploting the scatter plot of  category sepal for better visualisation
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],marker='+',color= 'red')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],marker='.',color= 'green')
```

```python
# ploting the scatter plot of  category petal for better visualisation
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],marker='+',color= 'red')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],marker='.',color= 'green')
```

```python
from sklearn.model_selection import train_test_split # importing training library
```

```python
x = df.drop(['target','flower_name'],axis ='columns') # splitting the data
x.head()
```

```python
y = df.target # defining the target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) # spliting the data in train and test category
```

```python
#importing model Support Vector Machine
from sklearn.svm import SVC
#model = SVC()
#model = SVC(C=1000)
model = SVC(C=100,gamma=5)
#model = SVC(kernel='linear')
```

```python
model.fit(x_train, y_train) # training the model
```

```python
model.score(x_test,y_test)  # to get the result
```
